```python
from Bio import Entrez
```

```python
Entrez.email = "shonushka.sen@berkeley.edu"
#I have chosen the following enzymes:
#For glycolysis: hexokinase, phosphofructokinase, phosphoglycerate kinase, and pyruvate kinase. 
#For the citric acid cycle: isocitrate dehydrogenase, Î±-Ketoglutarate dehydrogenase, succinyl-CoA synthetase
 #and succinate dehydrogenase. 
#For the pentose phosphate pathway: glucose-6-phosphate dehydrogenase, gluconolactonase, 
 #6-phosphogluconate dehydrogenase, and ribulose-5-phosphate isomerase. 
```

```python
handle = Entrez.esearch(db = 'nucleotide',
                       term = 'homo sapiens [ORGN] HXK1',
                       sort = 'relevance',
                       idtype = 'acc')
for i in Entrez.read(handle)['IdList']:
    handle = Entrez.efetch(db = 'nucleotide', id = i, rettype = 'fasta', retmode = 'text')
    print(handle.read())
#H. sapiens hexokinase (glycolysis) gene info found at https://www.ncbi.nlm.nih.gov/gene/3098
```

```python
handle = Entrez.esearch(db = 'nucleotide',
                       term = 'drosophila melanogaster [ORGN] HEX-T2',
                       sort = 'relevance',
                       idtype = 'acc')
for i in Entrez.read(handle)['IdList']:
    handle = Entrez.efetch(db = 'nucleotide', id = i, rettype = 'fasta', retmode = 'text')
    print(handle.read())
#D. melanogaster hexokinase (glycolysis) gene info found at https://www.ncbi.nlm.nih.gov/gene/43191
```
